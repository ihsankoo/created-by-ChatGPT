```python
import random

# The board is represented as a list of strings. The string values are:
# " ", an empty space; "X", an X; and "O", an O
board = [" ", " ", " ",
         " ", " ", " ",
         " ", " ", " "]

# The winning combinations
winning_combinations = [
    [0, 1, 2], [3, 4, 5], [6, 7, 8],  # rows
    [0, 3, 6], [1, 4, 7], [2, 5, 8],  # columns
    [0, 4, 8], [2, 4, 6]  # diagonals
]

# The players. 1 is X and -1 is O
players = [1, -1]

# The current player
player = 1

# The main game loop
while True:
    # Print the board
    print(f" {board[0]} | {board[1]} | {board[2]} ")
    print("-----------")
    print(f" {board[3]} | {board[4]} | {board[5]} ")
    print("-----------")
    print(f" {board[6]} | {board[7]} | {board[8]} ")
    print()

    # Get the player's move
    move = input("Enter your move (1-9): ")
    move = int(move) - 1

    # Make sure the move is valid
    if board[move] != " ":
        print("That space is already occupied. Please try again.")
        continue

    # Update the board
    board[move] = "X" if player == 1 else "O"

    # Check if the player has won
    for combination in winning_combinations:
        if board[combination[0]] == board[combination[1]] == board[combination[2]] and board[combination[0]] != " ":
            print(f"Player {player} wins!")
            exit()

    # Switch players
    player *= -1
```
